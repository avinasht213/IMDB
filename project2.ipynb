Whose director made the most movies?

```python
cur.execute('SELECT name, COUNT(director_id) FROM movies JOIN directors ON directors.id = movies.director_id GROUP BY director_id ORDER BY COUNT(director_id) DESC')
d=cur.fetchall()
d=pd.DataFrame(d, columns = ['director_name', 'movies_count'])
d.head(10)
```

Which movie(s) were directed by Brenda Chapman?

```python
cur.execute('select movies.original_title from movies inner join directors on movies.director_id=directors.id where directors.name="Brenda Chapman"')
d=cur.fetchall()
d=pd.DataFrame(d,columns=['movie_name'])
d
```

 Whose director is the most bankable?

```python
cur.execute('SELECT directors.name from directors inner join movies on directors.id=movies.director_id order by movies.revenue-movies.budget desc limit 1')
d=cur.fetchall()
d=pd.DataFrame(d,columns=['movie_name'])
d
```

Find the name of the 10th first women directors?
